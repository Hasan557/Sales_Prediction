```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_location = "C:/Msc course/Sem 2/data_analytics/Cw/sales data-set.csv"
col_list=["Date","Weekly_Sales"]
df = pd.read_csv(file_location)
df["year"] = pd.DatetimeIndex(df["Date"]).year
df['Weekly_Sales'] = df[(df['Weekly_Sales']>0)]['Weekly_Sales']
```

```python
grp = df.groupby('year').agg({"Weekly_Sales": [min, max, sum]})
print(grp)
```

```python
bp=df.boxplot(column='Weekly_Sales', by='year')
plt.show()
```

```python
bplot = sns.boxplot(y='Weekly_Sales', x='year', 
                 data=df, 
                 width=0.5,
                 palette="colorblind", showfliers=False) #without outliers
bplot.axes.set_title("Sales Data by year",
                    fontsize=16)
 
bplot.set_xlabel("Year", 
                fontsize=14)
 
bplot.set_ylabel("Weekly Total Sales",
                fontsize=14)
 
bplot.tick_params(labelsize=10)
```
